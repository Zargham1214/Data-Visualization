```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Read the file into a variable insurance_data
insurance_data = pd.read_csv(r"C:\Users\zargham abbas\Documents\Kaggle\Data Visualization\insurance.csv")
```

```python
insurance_data.head()
```

```python
plt.figure(figsize=(20,8))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])

#it shows a positive correlation between Body Mass Index (BMI) and charges 
```

```python
#to verify the 'positive correlation we can draw a regression line that best fits the data points'
plt.figure(figsize=(20,8))

sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#we can also display the relationship between more than two variables with colored scatter plot
plt.figure(figsize=(20,8))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['region'])
#this plot shows that smokers with increasing bmi pay more than non smokers with increasing bmi
```

```python
#fact discussed above can be verified with lmplot
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)

#In this plot regression line for smokers has a much steeper slope, relative to the line for nonsmokers!
```

```python
#the scatter plot to feature a categorical variable (like "smoker") on one of the main axes. 
#We'll refer to this plot type as a categorical scatter plot, and we build it with the sns.swarmplot command.
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])

#this plot shows
  # - on average non smokers pay less than smokers
  # - customers who pay the most charges are smokers
```
